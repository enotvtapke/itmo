```python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy.stats import boxcox
```

```python
df = pd.DataFrame(yf.download('AAPL', '2006-01-01', '2021-01-01').resample("MS").mean().Close)

df.plot()
df.head()
```

```python
df.apply(np.log).plot()
```

```python
boxcox(df, -0.2).plot()
```

```python
diff = df.diff().fillna(method='bfill')
diff.plot()
diff
```

```python
df = df.apply(np.log)
df.plot()
diff = df.diff().fillna(method='bfill')
diff.plot()
diff
```
