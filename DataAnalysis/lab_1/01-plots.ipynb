```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/SpotifyFeatures.csv')
print(data.head())
```

```python
plt.hist(data.popularity, bins=20)
plt.title("Popularity hist")
plt.show()
plt.boxplot(data.popularity)
plt.title("Popularity box plot")
plt.show()
```

```python
plt.hist2d(data.duration_ms,
           data.popularity,
           bins=50
          )
plt.show()

# No very popular tracks with long duration
```

```python
plt.scatter(data.popularity,
            data.genre,
            s=3,
           )
```

```python
plt.scatter(data.valence,
            data.danceability,
            s=1,
           )
plt.show()
plt.hist2d(data.valence,
           data.danceability,
           bins=30
          )
plt.show()

# Valence and danceability correlate with each other
```

```python
data.insert(data.columns.size, 'duration_min', np.round(data.duration_ms / 60000, 1))
```
