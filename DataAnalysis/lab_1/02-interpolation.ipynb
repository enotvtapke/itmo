```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

## Scipy Linear Interpolation

```python
n = 10
x = np.linspace(0, 5, n)
y = 2 * np.square(x - 2) + np.random.normal(0, 0.05, n)
f = scipy.interpolate.interp1d(x, y, kind='linear')
x_new = np.linspace(0, 5, 100)
y_new = f(x_new)

plt.plot(x, y, 'o', x_new, y_new, '-')
plt.show()
```

```python
import random
import interpolation

n = 10
x = random.sample(range(20), n)
x.sort()
y = np.random.normal(5, 5, n)
```

## Handmade Baricentric Lagrange Interpolation
Based on [Berrut and Trefethen 2004, “Barycentric Lagrange Interpolation”](https://people.maths.ox.ac.uk/trefethen/barycentric.pdf)

```python
f = interpolation.barycentric_interpolation(x, y)

x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x, y, 'o')
plt.plot(x, y, 'o', x_new, y_new, '-')
plt.show()
```

## Scipy Baricentric Lagrange Interpolation
See [scipy.interpolate.BarycentricInterpolator](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.BarycentricInterpolator.html#scipy.interpolate.BarycentricInterpolator)

```python
f = scipy.interpolate.BarycentricInterpolator(x, y)

x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x, y, 'o')
plt.plot(x, y, 'o', x_new, y_new, '-')
plt.show()
```
